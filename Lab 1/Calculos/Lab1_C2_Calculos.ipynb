```python
# Definición de los valores de las resistencias
R1 = 100
R2 = 10000
R3 = 1000
R4 = 100000
a = 0.05

# Listas de valores de RL y Vin para variar
valores_RL = [0.000000000001, 1000, 2000, 5000, 10000]
valores_Vin = [-1, 0.5, 2]

# Bucle anidado para recorrer los valores de RL y Vin
for Vin in valores_Vin:
    for RL in valores_RL:
        # Cálculo de I_RL
        I_RL = Vin / (R3 + a*RL) * 1000 * 1000
        # Impresión de los resultados
        print(f"RL = {RL} Ω Vin = {Vin} V I_RL= {I_RL:.7f} mA")
```
